```python
import pandas as pd
```

```python
epl = pd.read_csv('../data/raw/arsenal_epl_2025_26.csv')
ucl = pd.read_csv('../data/raw/arsenal_ucl_2025_26.csv')

# Ensure column names are clean before filtering
epl = epl.rename(columns=lambda x: x.strip() if isinstance(x, str) else x)
ucl = ucl.rename(columns=lambda x: x.strip() if isinstance(x, str) else x)
```

```python
# Check for upcoming fixtures (no result yet)
print("EPL - Upcoming fixtures:")
epl_upcoming = epl[epl['Result'].isna() | (ucl['Result'] == '') | (epl['Result'] == 'TBD')]
print(f"Count: {len(epl_upcoming)}")
print(epl_upcoming[['Date', 'Home Team', 'Away Team', 'Result']].head())
```

```python
print("\n" + "="*50)
print("UCL - Upcoming fixtures:")
ucl_upcoming = ucl[ucl['Result'].isna() | (ucl['Result'] == '') | (ucl['Result'] == 'TBD')]
print(f"Count: {len(ucl_upcoming)}")
print(ucl_upcoming[['Date', 'Home Team', 'Away Team', 'Result']].head())
```

```python

# Check current points from completed matches
print("\n" + "="*50)
print("EPL - Completed matches:")
epl_completed = epl[epl['Result'].notna() & (epl['Result'] != '') & (epl['Result'] != 'TBD')]
print(f"Count: {len(epl_completed)}")
```

```python
# Calculate current Arsenal points
def get_arsenal_result(row):
    if pd.isna(row['Result']) or row['Result'] == '':
        return None
    
    home_goals, away_goals = map(int, row['Result'].split(' - '))
    
    if row['Home Team'] == 'Arsenal':
        if home_goals > away_goals:
            return 3
        elif home_goals == away_goals:
            return 1
        else:
            return 0
    else:  # Arsenal is away
        if away_goals > home_goals:
            return 3
        elif home_goals == away_goals:
            return 1
        else:
            return 0
```

```python
epl_completed['arsenal_points'] = epl_completed.apply(get_arsenal_result, axis=1)
current_points = epl_completed['arsenal_points'].sum()
```

```python
print(f"\nArsenal current points: {current_points}")
print(f"Matches played: {len(epl_completed)}")
print(f"Matches remaining: {len(epl_upcoming)}")
```
