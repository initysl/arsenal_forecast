```python
import pandas as pd

df = pd.read_csv('../data/raw/epl_final.csv')
```

```python
# Get last 3 seasons
seasons = df['Season'].unique()
print(f"Available seasons: {sorted(seasons)}")

last_3_seasons = sorted(seasons)[-3:]  # Takes last 3
print(f"Using: {last_3_seasons}")
```

```python
# Filter
df_filtered = df[df['Season'].isin(last_3_seasons)]

# Now filter Arsenal
arsenal = df_filtered[
    (df_filtered['HomeTeam'] == 'Arsenal') | 
    (df_filtered['AwayTeam'] == 'Arsenal')
].copy()

print(f"\nArsenal matches in last 3 seasons: {len(arsenal)}")
print(arsenal[['Season', 'MatchDate', 'HomeTeam', 'AwayTeam', 
               'FullTimeHomeGoals', 'FullTimeAwayGoals']].head(10))
```
