```python
import pandas as pd
```

```python
epl = pd.read_csv('../data/raw/mancity_epl_2025_26.csv')

# Ensure column names are clean before filtering
epl = epl.rename(columns=lambda x: x.strip() if isinstance(x, str) else x)
```

```python
# Check for upcoming fixtures (no result yet)
print("EPL - Upcoming fixtures:")
epl_upcoming = epl[epl['Result'].isna() | (epl['Result'] == '') | (epl['Result'] == 'TBD')]
print(f"Count: {len(epl_upcoming)}")
print(epl_upcoming[['Date', 'Home Team', 'Away Team', 'Result']].head())
```

```python

# Check current points from completed matches
print("\n" + "="*50)
print("EPL - Completed matches:")
epl_completed = epl[epl['Result'].notna() & (epl['Result'] != '') & (epl['Result'] != 'TBD')]
print(f"Count: {len(epl_completed)}")
```

```python
# Calculate current Man City points
def get_mancity_result(row):
    if pd.isna(row['Result']) or row['Result'] == '':
        return None
    
    home_goals, away_goals = map(int, row['Result'].split(' - '))
    
    if row['Home Team'] == 'Man City':
        if home_goals > away_goals:
            return 3
        elif home_goals == away_goals:
            return 1
        else:
            return 0
    else:  # Man City is away
        if away_goals > home_goals:
            return 3
        elif home_goals == away_goals:
            return 1
        else:
            return 0
```

```python
epl_completed['mancity_points'] = epl_completed.apply(get_mancity_result, axis=1)
current_points = epl_completed['mancity_points'].sum()
```

```python
print(f"\nMan City current points: {current_points}")
print(f"Matches played: {len(epl_completed)}")
print(f"Matches remaining: {len(epl_upcoming)}")
```
